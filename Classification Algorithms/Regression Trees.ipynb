```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```
df = pd.read_csv('real_estate_data.csv')
df.head()
```

```
df.shape
```

```
df.isna().sum()
```

```
df.dropna(inplace=True)
df.isna().sum()
```

```
df.shape
```

```
X = df.drop(columns=["MEDV"])
Y = df["MEDV"]
```

```
X.head()
```

```
Y.head()
```

```
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```
DecRegr = DecisionTreeRegressor(criterion = 'squared_error')
DecRegr.fit(x_train, y_train)
```

```
DecRegr.score(x_test, y_test)
```

```
y_pred = DecRegr.predict(x_test)
print("$",(y_pred - y_test).abs().mean()*1000)
```

```
DecRegr = DecisionTreeRegressor(criterion = 'absolute_error')
DecRegr.fit(x_train, y_train)
print(DecRegr.score(x_test, y_test))
y_pred = DecRegr.predict(x_test)
print("$",(y_pred - y_test).abs().mean()*1000)
```
