```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate data
np.random.seed(0)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 2 * X**2 + 3 * X + 1 + np.random.randn(100, 1) * 3  # Quadratic with noise

# Plot the original data
plt.scatter(X, y, color='blue', label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data with Underlying Quadratic Pattern')
plt.legend()
plt.show()

# Underfitting: Linear model
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_linear, color='red', label='Linear fit (Underfitting)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Underfitting Example')
plt.legend()
plt.show()

# Overfitting: High-degree polynomial model
poly_high_degree = make_pipeline(PolynomialFeatures(degree=10), LinearRegression())
poly_high_degree.fit(X, y)
y_pred_high_degree = poly_high_degree.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_high_degree, color='red', label='High-degree Polynomial fit (Overfitting)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Overfitting Example')
plt.legend()
plt.show()

# Appropriate fit: Quadratic model
poly_quadratic = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
poly_quadratic.fit(X, y)
y_pred_quadratic = poly_quadratic.predict(X)

plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred_quadratic, color='green', label='Quadratic fit (Appropriate fit)')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Appropriate Fit Example')
plt.legend()
plt.show()
```
