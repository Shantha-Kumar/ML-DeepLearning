```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
import sklearn.tree as tree
```

```
df = pd.read_csv('drug200.csv', delimiter=',')
df.head()
```

```
print("Length of the Data ", len(df))
print("Shape ", df.shape)
```

```
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].to_numpy()
X[0:5]
```

```
df['Drug'].value_counts()
```

```
df[['Sex', 'BP']].value_counts()
```

### Converting Categorical Values to Numerical or Dummy values

```
from sklearn import preprocessing

le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3]) 

X[0:5]
```

```
Y = df['Drug'].to_numpy()
Y[0:5]
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=3)
```

```
print("X Train shape", x_train.shape)
print("Y Train shape", y_train.shape)
```

```
print("X Test shape", x_test.shape)
print("Y Test shape",y_test.shape)
```

```
drugclf = DecisionTreeClassifier(criterion="entropy", max_depth=4)
drugclf
```

```
drugclf.fit(x_train, y_train)
```

```
y_pred = drugclf.predict(x_test)
```

```
print("y_pred")
print(y_pred)
print("y_test")
print(y_test)
```

```
from sklearn import metrics
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```
