```
weights = np.around(np.random.uniform(size=6), decimals=2)
biases = np.around(np.random.uniform(size=3), decimals=2)
print(weights)
print(biases)
```

```
x_1 = 0.5
x_2 = 0.85
print(f'x1 is {x_1} and x2 is {x_2}')
```

```
z_11 = x_1 * weights[0] + x_2 * weights[1] + biases[0]
print(f"The weighted sum of the inputs at the first node in the hidden layer is {np.around(z_11, decimals=4)}")

z_12 = x_1 * weights[2] + x_2 * weights[3] + biases[1]
print(f"The weighted sum of the inputs at the second node in the hidden layer is {np.around(z_12, decimals=4)}")
```

```
a_11 = 1.0 / (1.0 + np.exp(-z_11))
print(f'The activation of the first node in the hidden layer is {np.around(a_11, decimals=4)}')

a_12 = 1.0 / (1.0 + np.exp(-z_12))
print(f'The activation of the first node in the hidden layer is {np.around(a_12, decimals=4)}')
```

```
z_2 = a_11 * weights[4] + a_12 * weights[5] + biases[2]
print(f"The weighted sum of the inputs at the node in the final layer is {np.around(z_2, decimals=4)}")
```

```
a_2 = 1.0 / (1.0 + np.exp(-z_2))
print(f'The output of the network for x1 = 0.5 and x2 = 0.85 is {np.around(a_2, decimals=4)}')
```
